# **VIF check**

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler


X = df.drop('booking_complete', axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns

vif_data["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]

print(vif_data)
```

# **Training the model**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_auc_score, accuracy_score, confusion_matrix

# Define the features and the target
X = df.drop('booking_complete', axis=1)
y = df['booking_complete']

# Train a Random Forest classifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Get the predicted probabilities for the positive class (1)
y_probs = clf.predict_proba(X_test)[:, 1]
y_pred = clf.predict(X_test)

# Sort true labels and predicted probabilities by predicted probability
sorted_indices = np.argsort(y_probs)
y_test_sorted = np.array(y_test)[sorted_indices]
y_probs_sorted = y_probs[sorted_indices]

# Calculate cumulative gains (Lorenz curve)
cumulative_gains = np.cumsum(y_test_sorted) / sum(y_test_sorted)
cumulative_random = np.linspace(0, 1, len(y_test_sorted))

# Plot the Lorenz curve (Gini curve)
plt.figure(figsize=(8, 6))
plt.plot(cumulative_random, cumulative_gains, label='Gini Curve', color='blue', lw=2)
plt.plot([0, 1], [0, 1], label='Random Model', linestyle='--', color='gray')
plt.fill_between(cumulative_random, cumulative_random, cumulative_gains, color='lightblue', alpha=0.6)

# Gini coefficient
gini_coefficient = 2 * roc_auc_score(y_test, y_probs) - 1
plt.text(0.6, 0.2, f'Gini Coefficient: {gini_coefficient:.2f}', fontsize=12, color='blue')

# Labels and title
plt.title('Gini Curve (Lorenz Curve)')
plt.xlabel('Proportion of Samples')
plt.ylabel('Cumulative True Positives')
plt.legend()
plt.grid(True)
plt.show()

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

```

# *Cross-Validation*

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, f1_score, roc_auc_score
import numpy as np


kf = KFold(n_splits=5, shuffle=True, random_state=42)

f1_scorer = make_scorer(f1_score)
gini_scorer = make_scorer(lambda y_true, y_pred: 2 * roc_auc_score(y_true, y_pred) - 1)

accuracy_scores = cross_val_score(clf, X_resampled, y_resampled, cv=kf, scoring='accuracy')
f1_scores = cross_val_score(clf, X_resampled, y_resampled, cv=kf, scoring=f1_scorer)
gini_scores = cross_val_score(clf, X_resampled, y_resampled, cv=kf, scoring=gini_scorer)

print(f"Cross-Validation Accuracy Scores for each fold: {accuracy_scores}")
print(f"Mean Accuracy: {np.mean(accuracy_scores):.4f}")
print(f"Standard Deviation of Accuracy: {np.std(accuracy_scores):.4f}")

print(f"Cross-Validation F1 Scores for each fold: {f1_scores}")
print(f"Mean F1 Score: {np.mean(f1_scores):.4f}")
print(f"Standard Deviation of F1 Score: {np.std(f1_scores):.4f}")

print(f"Cross-Validation Gini Scores for each fold: {gini_scores}")
print(f"Mean Gini Score: {np.mean(gini_scores):.4f}")
print(f"Standard Deviation of Gini Score: {np.std(gini_scores):.4f}")
```
